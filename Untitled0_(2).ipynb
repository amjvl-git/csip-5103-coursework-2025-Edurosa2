```python
# Imports
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split

# Transformation
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Load datasets
train_dataset = datasets.FashionMNIST(root='data', train=True, download=True, transform=transform)
test_dataset = datasets.FashionMNIST(root='data', train=False, download=True, transform=transform)

# Split datasets
train_size = int(0.8 * len(train_dataset))
val_size = len(train_dataset) - train_size
train_data, val_data = random_split(train_dataset, [train_size, val_size])

# DataLoaders
batch_size = 64
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Define ANN model
class SimpleANN(nn.Module):
    def __init__(self):
        super(SimpleANN, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(28*28, 128)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(128, 64)
        self.relu2 = nn.ReLU()
        self.output = nn.Linear(64, 10)

    def forward(self, x):
        x = self.flatten(x)
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        x = self.output(x)
        return x

# Instantiate model
ann_model = SimpleANN()

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
ann_model.to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(ann_model.parameters(), lr=0.001)

# Training loop
epochs = 10
train_losses = []
val_losses = []

for epoch in range(epochs):
    ann_model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = ann_model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    # Validation
    ann_model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = ann_model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

    avg_train_loss = running_loss / len(train_loader)
    avg_val_loss = val_loss / len(val_loader)
    train_losses.append(avg_train_loss)
    val_losses.append(avg_val_loss)

    print(f"Epoch [{epoch+1}/{epochs}] - Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")

print("✅ Training complete.")

# 🔽 Save the trained ANN model weights
torch.save(ann_model.state_dict(), "ann_fmnist.pt")
print("💾 ANN model saved to ann_fmnist.pt")

```

```python
# ==== CNN: define, train (best-on-val), save, and evaluate ====

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay

# ✅  define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ✅ Class names (for confusion matrix)
class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
               'Sandal','Shirt','Sneaker','Bag','Ankle boot']

# --- Model definition ---
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, 3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25),
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64*7*7, 128), nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 10),
        )
    def forward(self, x):
        return self.classifier(self.features(x))

# --- Helper functions ---
def accuracy_on_loader(model, loader, device):
    model.eval(); correct = total = 0
    with torch.no_grad():
        for x,y in loader:
            x,y = x.to(device), y.to(device)
            pred = model(x).argmax(1)
            correct += (pred==y).sum().item()
            total   += y.size(0)
    return correct/total if total>0 else 0.0

def evaluate_model(model, loader, device):
    model.eval(); ys=[]; ps=[]
    with torch.no_grad():
        for x,y in loader:
            x = x.to(device)
            p = model(x).argmax(1).cpu().numpy()
            ys.append(y.numpy()); ps.append(p)
    y = np.concatenate(ys); p = np.concatenate(ps)
    acc = (y==p).mean()
    return acc, y, p

# --- Train ---
cnn_model = SimpleCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(cnn_model.parameters(), lr=1e-3)

epochs = 10
best_val_acc, best_state = 0.0, None
train_losses, val_losses = [], []
train_acc_hist, val_acc_hist = [], []

for ep in range(epochs):
    cnn_model.train(); run_loss = 0.0
    for x,y in train_loader:
        x,y = x.to(device), y.to(device)
        optimizer.zero_grad()
        loss = criterion(cnn_model(x), y)
        loss.backward(); optimizer.step()
        run_loss += loss.item()
    avg_tr = run_loss/len(train_loader)

    # val
    cnn_model.eval(); vloss=0.0
    with torch.no_grad():
        for x,y in val_loader:
            x,y = x.to(device), y.to(device)
            vloss += criterion(cnn_model(x), y).item()
    avg_va = vloss/len(val_loader)

    tr_acc = accuracy_on_loader(cnn_model, train_loader, device)
    va_acc = accuracy_on_loader(cnn_model, val_loader, device)
    if va_acc > best_val_acc:
        best_val_acc = va_acc
        best_state = {k: v.cpu().clone() for k,v in cnn_model.state_dict().items()}

    train_losses.append(avg_tr); val_losses.append(avg_va)
    train_acc_hist.append(tr_acc); val_acc_hist.append(va_acc)
    print(f"[CNN {ep+1}/{epochs}] Train {avg_tr:.4f} | Val {avg_va:.4f} | "
          f"Train Acc {tr_acc:.4f} | Val Acc {va_acc:.4f}")

# restore best + save
if best_state: cnn_model.load_state_dict(best_state)
torch.save(cnn_model.state_dict(), "cnn_fmnist.pt")
print(f"💾 Saved best CNN to cnn_fmnist.pt (best Val Acc = {best_val_acc:.4f})")

# --- Evaluate on test ---
test_acc, y_test, p_test = evaluate_model(cnn_model, test_loader, device)
print(f"✅ CNN — Test Accuracy: {test_acc:.4f}")

# Confusion matrix + report
cm = confusion_matrix(y_test, p_test)
fig, ax = plt.subplots(figsize=(7,7))
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)\
    .plot(ax=ax, xticks_rotation=45, colorbar=False)
plt.title("CNN — Test Confusion Matrix"); plt.tight_layout(); plt.show()

print("CNN — Classification Report")
print(classification_report(y_test, p_test, target_names=class_names, digits=3))

# Training curves
plt.figure(); plt.plot(train_losses, label="Train"); plt.plot(val_losses, label="Val")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("CNN — Loss Curves"); plt.legend(); plt.show()

plt.figure(); plt.plot(train_acc_hist, label="Train"); plt.plot(val_acc_hist, label="Val")
plt.xlabel("Epoch"); plt.ylabel("Accuracy"); plt.title("CNN — Accuracy Curves"); plt.legend(); plt.show()
```

```python
# ==== FINAL ALL-IN-ONE CNN CELL ====

# Imports
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import pandas as pd

# ✅ Define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# Class names
class_names = ['T-shirt/top','Trouser','Pullover','Dress','Coat',
               'Sandal','Shirt','Sneaker','Bag','Ankle boot']

# Helper: accuracy
def accuracy_on_loader(model, loader, device):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            pred = model(x).argmax(1)
            correct += (pred == y).sum().item()
            total += y.size(0)
    return correct / total if total > 0 else 0.0

# Helper: evaluate model (for confusion matrix + accuracy)
def evaluate_model(model, data_loader, device):
    model.eval()
    all_y, all_pred = [], []
    with torch.no_grad():
        for x, y in data_loader:
            x = x.to(device)
            pred = model(x).argmax(1).cpu().numpy()
            all_y.extend(y.numpy())
            all_pred.extend(pred)
    y = np.array(all_y); p = np.array(all_pred)
    acc = (y == p).mean()
    return acc, y, p

# ==== CNN Definition ====
class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Dropout(0.25)
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64*7*7, 128), nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, 10)
        )
    def forward(self, x):
        return self.classifier(self.features(x))

# ==== Model, loss, optimizer ====
cnn_model = SimpleCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(cnn_model.parameters(), lr=1e-3)

# ==== Training (with best-on-validation) ====
epochs_cnn = 10
cnn_train_losses, cnn_val_losses = [], []
cnn_train_acc, cnn_val_acc_hist = [], []  # rename val history to avoid overwriting val_acc variable below
best_val_acc, best_state = 0.0, None

for epoch in range(epochs_cnn):
    cnn_model.train()
    run_loss = 0.0
    for x, y in train_loader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()
        logits = cnn_model(x)
        loss = criterion(logits, y)
        loss.backward()
        optimizer.step()
        run_loss += loss.item()

    # Validation
    cnn_model.eval()
    vloss = 0.0
    with torch.no_grad():
        for x, y in val_loader:
            x, y = x.to(device), y.to(device)
            vloss += criterion(cnn_model(x), y).item()

    avg_tr = run_loss / len(train_loader)
    avg_va = vloss / len(val_loader)
    acc_tr = accuracy_on_loader(cnn_model, train_loader, device)
    acc_va = accuracy_on_loader(cnn_model, val_loader, device)

    cnn_train_losses.append(avg_tr)
    cnn_val_losses.append(avg_va)
    cnn_train_acc.append(acc_tr)
    cnn_val_acc_hist.append(acc_va)

    # ✅ track best checkpoint by validation accuracy
    if acc_va > best_val_acc:
        best_val_acc = acc_va
        best_state = {k: v.detach().cpu().clone() for k, v in cnn_model.state_dict().items()}

    print(f"[CNN {epoch+1}/{epochs_cnn}] "
          f"Train Loss {avg_tr:.4f} | Val Loss {avg_va:.4f} | "
          f"Train Acc {acc_tr:.4f} | Val Acc {acc_va:.4f}")

# ✅ Restore best weights and save
if best_state is not None:
    cnn_model.load_state_dict(best_state)
torch.save(cnn_model.state_dict(), "cnn_fmnist.pt")
print(f"💾 CNN model saved to cnn_fmnist.pt (best Val Acc = {best_val_acc:.4f})")

# ==== Final test evaluation ====
cnn_test_acc, cnn_test_y, cnn_test_pred = evaluate_model(cnn_model, test_loader, device)
print(f"\n✅ CNN — Test Accuracy: {cnn_test_acc:.4f}")

# ==== Confusion Matrix ====
cm_cnn = confusion_matrix(cnn_test_y, cnn_test_pred)
fig, ax = plt.subplots(figsize=(7,7))
ConfusionMatrixDisplay(confusion_matrix=cm_cnn, display_labels=class_names).plot(
    ax=ax, xticks_rotation=45, colorbar=False)
plt.title("CNN — Test Confusion Matrix")
plt.tight_layout()
plt.show()

# ==== Classification Report ====
print("CNN — Classification Report:")
print(classification_report(cnn_test_y, cnn_test_pred, target_names=class_names, digits=3))

# ==== Plot training curves ====
plt.figure()
plt.plot(cnn_train_losses, label="Train")
plt.plot(cnn_val_losses, label="Val")
plt.xlabel("Epoch"); plt.ylabel("Loss")
plt.title("CNN — Loss Curves")
plt.legend(); plt.show()

plt.figure()
plt.plot(cnn_train_acc, label="Train")
plt.plot(cnn_val_acc_hist, label="Val")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.title("CNN — Accuracy Curves")
plt.legend(); plt.show()

# ==== Compare ANN and CNN performance ====
ann_val_acc  = accuracy_on_loader(ann_model, val_loader, device)
ann_test_acc = accuracy_on_loader(ann_model, test_loader, device)
cnn_val_acc  = accuracy_on_loader(cnn_model, val_loader, device)
cnn_test_acc = accuracy_on_loader(cnn_model, test_loader, device)

summary = pd.DataFrame({
    "Model": ["ANN", "CNN"],
    "Validation Accuracy": [ann_val_acc, cnn_val_acc],
    "Test Accuracy": [ann_test_acc, cnn_test_acc],
})
display(summary)
```
